```python
import cv2  # Open CV library for Computer Vision and Image Processing
import matplotlib.pyplot as plt 
import numpy as np
```

```python
image = cv2.imread('../input/image/1.jpg')
plt.imshow(image)
```

```python
image = cv2.imread('../input/image/1.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)    # Conversion of image Into Gray Color
plt.imshow(gray, cmap ='gray')
```

```python
blur =cv2.GaussianBlur(gray,(11,11),0)  
plt.imshow(blur, cmap='gray')
```

```python
canny = cv2.Canny(blur,30,105,3)    # Using Canny Edge Detection Algorithm
plt.imshow(canny, cmap='gray')
```

```python
canny
```

```python
dilated =cv2.dilate(canny,(1,1), iterations =2)    # Dilation is used to observe changes
plt.imshow(dilated, cmap='gray')
```

```python
(cnt, heirarchy ) = cv2.findContours(dilated.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
rgb =cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2.drawContours(rgb, cnt, -1, (0,255,0),2)
```

```python
plt.imshow(rgb)
```

```python
print('Cells in the image', len(cnt))   # It outputs the object count
```
